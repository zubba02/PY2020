```python
#IMPORT LIBRRARIES
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
#RAIN

Hanimaadhoo_Rains = pd.read_csv('Hanimaadhoo Data Rain.csv',infer_datetime_format=True)
#Hanimaadhoo_Rains['Years & Months'] =  pd.to_datetime(Hanimaadhoo_Speeds['Years & Months']) # This parses to date time.
Hanimaadhoo_Rains2 = Hanimaadhoo_Rains.replace(['TR','NIL','TR '], [0.1,0.0,0.1]).fillna(value='-99999')

dates = pd.date_range(start='6/1/1991', end='12/31/2020')
#print(dates.strftime('%Y-%m-%d'))
```

```python
rain_dates = []
rains = []

for i in dates:
    #print (i.strftime('%Y-%m-%d'))
    rain_dates.append(i)
    orig_date = i.strftime('%Y-%m')
    for index, row in Hanimaadhoo_Rains2.iterrows():
        if orig_date == row['Years & Months']:
            d = int(i.strftime('%d'))
            print (row['Day {}'.format(d)])
            rains.append(row['Day {}'.format(d)])
            
```

```python
np.savetxt('Rains.csv', np.c_[rain_dates,rains], delimiter=',', fmt='%s') 

all_rains = np.c_[rain_dates,rains]

rains_df = pd.DataFrame(all_rains, columns = ['rain_dates','rains'])

rains_df.to_pickle('rains.pkl')
```

```python
rains_df['rains'] = rains_df['rains'].astype(float)
summed_monthly_annual = rains_df.groupby([rains_df['rain_dates'].dt.year.rename('year'), rains_df['rain_dates'].dt.month_name().rename('month')])['rains'].sum().reset_index()
```

```python
monthly_annual = summed_monthly_annual.groupby(summed_monthly_annual['month']).mean()
monthly_annual
monthly_annual.to_pickle('monthly_annual.pkl')
monthly_annual
```

```python
### GIVE A TRY FOR NUMBER OF RAINY DAYS ON AVERAGE ANNUALLY

# light rain  : > 2.5mm
# heavy rain : 2.5 mm - 7.6 mm
# heavy rain : < 7.6 mm


rain_date = pd.date_range(start='6/1/1991', end='12/31/2020')

rains_dates = []
light_rain = []
moderate_rain = []
heavy_rain = []
mod_heavy_rain = []


for i in rain_date:
    print (i.strftime('%Y-%m-%d'))
    rains_dates.append(i)
    orig_date = i.strftime('%Y-%m')
    for index, row in Hanimaadhoo_Rains2.iterrows():
        if orig_date == row['Years & Months']:
            d = int(i.strftime('%d'))
            rain_mm = float(row['Day {}'.format(d)])
            if rain_mm < 2.5:
                print ('This is light rain')
                light_rain.append(1)
            else:
                light_rain.append(0)

                
for i in rain_date:
    print (i.strftime('%Y-%m-%d'))
    #rain_dates.append(i)
    orig_date = i.strftime('%Y-%m')
    for index, row in Hanimaadhoo_Rains2.iterrows():
        if orig_date == row['Years & Months']:
            d = int(i.strftime('%d'))
            rain_mm = float(row['Day {}'.format(d)])
            if rain_mm > 2.5 and rain_mm < 7.6:
                print ('This is moderate_rain')
                moderate_rain.append(1)
            else:
                moderate_rain.append(0)
                
                
for i in rain_date:
    print (i.strftime('%Y-%m-%d'))
    #rain_dates.append(i)
    orig_date = i.strftime('%Y-%m')
    for index, row in Hanimaadhoo_Rains2.iterrows():
        if orig_date == row['Years & Months']:
            d = int(i.strftime('%d'))
            rain_mm = float(row['Day {}'.format(d)])
            if rain_mm > 7.6:
                print ('This is heavy_rain')
                heavy_rain.append(1)
            else:
                heavy_rain.append(0)   
                
                
                
for i in rain_date:
    print (i.strftime('%Y-%m-%d'))
    #rain_dates.append(i)
    orig_date = i.strftime('%Y-%m')
    for index, row in Hanimaadhoo_Rains2.iterrows():
        if orig_date == row['Years & Months']:
            d = int(i.strftime('%d'))
            rain_mm = float(row['Day {}'.format(d)])
            if rain_mm > 2.5:
                print ('This is Mod_heavy_rain')
                mod_heavy_rain.append(1)
            else:
                mod_heavy_rain.append(0) 
```

```python
np.savetxt('Rain_Types.csv', np.c_[rains_dates,light_rain,moderate_rain,heavy_rain], delimiter=',', fmt='%s') 

rain_types = np.c_[rains_dates,light_rain,moderate_rain,heavy_rain,mod_heavy_rain]

raintp_df = pd.DataFrame(rain_types, columns = ['rain_dates','light_rain','moderate_rain','heavy_rain', 'mod_heavy_rain'])

raintp_df.to_pickle('rain_types.pkl')
```

```python
#raintp_df['rains'] = raintp_df['rains'].astype(float)
summed_monthly_annual_rainy_days = raintp_df.groupby([raintp_df['rain_dates'].dt.year.rename('year'), raintp_df['rain_dates'].dt.month_name().rename('month')])['mod_heavy_rain'].sum().reset_index()

#summed_monthly_annual_rainy_days

monthly_average_rain_ays = summed_monthly_annual_rainy_days.groupby(pd.PeriodIndex(summed_monthly_annual_rainy_days['month'], freq="M"))['mod_heavy_rain'].mean()


#monthly_average_rain_ays.insert(2, "Month", ['Jan', 'Feb', 'Mar', 'Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], True)
```
